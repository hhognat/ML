```python
from transformers import AutoTokenizer

model_name = "bert-base-uncased"  # Example model
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Get vocabulary size
vocab_size = tokenizer.vocab_size

# Get total text data size (example - reading a text file)
with open("hellaswag_prompt.txt", "r", encoding="utf-8") as f:
    text = f.read()
text_size_bytes = len(text.encode("utf-8"))  # Byte count
text_size_mb = text_size_bytes / (1024 * 1024) # Convert bytes to MB
text_size_mb = text_size_bytes / (1024) # Convert bytes to KB

# Calculate compression ratio
compression_ratio = text_size_mb / vocab_size

print(f"Vocabulary Size: {vocab_size}")
print(f"Total Text Size: {text_size_mb:.2f} MB")
print(f"Compression Ratio: {compression_ratio:.4f}")
```
